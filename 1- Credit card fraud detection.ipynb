### EDA steps-credit card fraud data

```python
import seaborn as crdcd_fdsea
import matplotlib.pyplot as crdcd_fdmt
import warnings as crdcd_fdwrin
crdcd_fdwrin.filterwarnings("ignore")
```

```python
crdcd_fdmt.figure(figsize=(18, 12))
crdcd_fdsea.heatmap(crdcd_fd.corr(), annot=True)
```

#### Countplot for credit card- 'default.payment.next.month'

```python
crdcd_fdsea.countplot(x ='default.payment.next.month', data = crdcd_fd, color='pink')
crdcd_fdmt.title("credit card 'default.payment.next.month'",size=10)
crdcd_fdmt.show()
```

(1=yes,0=no)
* Default credit card payements are less.

#### credit card member's sex and marriage status

```python
crdcd_fdsea.countplot(x ='SEX', hue = "MARRIAGE", data = crdcd_fd)
crdcd_fdmt.legend(loc='upper left', bbox_to_anchor=(1.2, .80))
crdcd_fdmt.title("credit card member's sex and marriage status'")
crdcd_fdmt.show()
```

1=married, 2=single, 3= others
*unmarried credit card members are high. In which Female unmarried members are majority compared to Males

#### Distplot for credit card members age details

```python
crdcd_fdmt.figure(figsize = (6, 4))
crdcd_fdsea.distplot(crdcd_fd['AGE'], color='red')
crdcd_fdmt.title("credit card members age details")
crdcd_fdmt.ylabel("count")
crdcd_fdmt.show()
```

Age of credit card members lies between 20 to 80

#### Credit card members age and limited balance

```python
crdcd_fdsea.scatterplot(data=crdcd_fd, x="LIMIT_BAL", y="AGE")
crdcd_fdmt.title("credit card members age and their limited balance details")
crdcd_fdmt.show()
```

Members those having age between 25 to 50, whose credit card limited balance amount is high.

```python
crdcd_fdsea.countplot(x ='SEX', data = crdcd_fd)
crdcd_fdmt.title("Sex Details")
crdcd_fdmt.show()
```

** 1=male, 2=female
** Female credit card holders are majority.

```python
crdcd_fdsea.countplot(x ='EDUCATION', data = crdcd_fd)
crdcd_fdmt.title("Education Details")
crdcd_fdmt.show()
```

** 1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown.
** university passed members are majority.

```python
crdcd_fdsea.countplot(x ='MARRIAGE', data = crdcd_fd, color='grey')
crdcd_fdmt.title("Marriage Details")
crdcd_fdmt.show()
```

** 1=married, 2=single, 3=others
** Unmarried credit card holders are majority.

**Repayment status for 7 months**

```python
for o in ['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']:
  vable = crdcd_fdsea.distplot(crdcd_fd[o], color ='green', bins = 25)
  vable.set(title =str(o)+ ' Repayment status')
  crdcd_fdmt.show()
```

**[-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, … 8=payment delay for eight months, 9=payment delay for nine months and above]
** so most of the members pays at correct time without any delay.

**Amount of bill statement for 6 months**

```python
for o in ['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6']:
  vable = crdcd_fdsea.distplot(crdcd_fd[o], color ='red', bins = 25)
  vable.set(title =str(o)+ '- Amount of bill statement')
  crdcd_fdmt.show()
```

** Statement shows bill amount of month september 2005 to April 2005.

**Amount of previous payment for 6 months**

```python
for o in ['PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']:
  vable = crdcd_fdsea.distplot(crdcd_fd[o], color ='tan', bins = 25)
  vable.set(title =str(o)+ '- Amount of previous payment')
  crdcd_fdmt.show()
```

** Statement shows Amount of previous payment in month september 2005 to April 2005.

```python
vable = crdcd_fdsea.boxplot(y ="LIMIT_BAL", x ="default.payment.next.month", data = crdcd_fd, palette = "Set2")

vable.set(ylabel ="balance limit", xlabel = "default payment", title ='defauly payment with balance limit')
```

** Default payment (1=yes, 0=no)
