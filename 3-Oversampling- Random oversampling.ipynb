### Random oversampling

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler as crdcd_fdrddm
from collections import Counter as crdcd_fdcnt
print('starting -crdcd_fd datashape %s' % crdcd_fdcnt(crdcd_fd_o_Y))
crd_mdl= crdcd_fdrddm()
crdcd_fd_i_X, crdcd_fd_o_Y = crd_mdl.fit_resample(crdcd_fd_i_X, crdcd_fd_o_Y)
print('After random oversampling -crdcd_fd datashape %s' % crdcd_fdcnt(crdcd_fd_o_Y))
```

```python
from sklearn.model_selection import train_test_split as crdcd_fdtrrn
from sklearn.metrics import classification_report as crdcd_fdclfrt
from sklearn.metrics import confusion_matrix as crdcd_fdcfssm
from sklearn.metrics import ConfusionMatrixDisplay as crdcd_fdcfxdp
from sklearn.model_selection import GridSearchCV as crdcd_fdgshc
import warnings as crdcd_fdwin
crdcd_fdwin.filterwarnings("ignore")
import time as crdcd_fdttee
```

```python
# train==60%
crdcd_fd_i_Xnt, crdcd_fd_i_Xst, crdcd_fd_o_Ynt, crdcd_fd_o_Yst = crdcd_fdtrrn(crdcd_fd_i_X, crdcd_fd_o_Y, test_size=0.4, random_state=33)
```

```python
# validation==20% & test==20%
crdcd_fd_i_Xdv, crdcd_fd_i_Xst, crdcd_fd_o_Ydv, crdcd_fd_o_Yst = crdcd_fdtrrn(crdcd_fd_i_Xst, crdcd_fd_o_Yst, test_size=0.5, random_state=33)
```

```python
print("credit card train   : ", crdcd_fd_i_Xnt.shape[0])
print("credit card validation: ", crdcd_fd_i_Xdv.shape[0])
print("credit card test : ", crdcd_fd_i_Xst.shape[0])
```

### ML algorithm-credit card fraud detection

#### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier as crdcd_fddssn

crd_par = { 'criterion':['gini','entropy','log_loss'],
              'splitter': ['best','random'],
               'min_samples_leaf': [1,2,3,4,5],
            'max_features':['auto','log2','sqrt']}


crd_par_bs = crdcd_fddssn(random_state=33)
crd_par_bs = crdcd_fdgshc(crd_par_bs, crd_par, cv=2, verbose=1)
crd_par_bs.fit(crdcd_fd_i_Xnt.sample(1000, random_state=33), crdcd_fd_o_Ynt.sample(1000, random_state=33))

print( crd_par_bs.best_params_)
print("credit card-score: ", crd_par_bs.best_score_)
```

```python
crdcd_pt1 = crdcd_fdttee.time()

crdcd_fd_MA= crdcd_fddssn(criterion= 'entropy', max_features= 'auto', min_samples_leaf=1 ,splitter='best' )
crdcd_fd_MA.fit(crdcd_fd_i_Xnt, crdcd_fd_o_Ynt)

crdcd_pt2 = crdcd_fdttee.time()
print("Duration for Training:", crdcd_pt2-crdcd_pt1)
```

```python
crdcd_pt1 = crdcd_fdttee.time()

crdcd_fd_Pp= crdcd_fd_MA.predict(crdcd_fd_i_Xdv)
print(crdcd_fdclfrt(crdcd_fd_o_Ydv, crdcd_fd_Pp))

crdcd_pt2 = crdcd_fdttee.time()
crdcd_rdo = crdcd_fdcfssm(crdcd_fd_o_Ydv, crdcd_fd_Pp)
crdcd_rex = crdcd_fdcfxdp(confusion_matrix = crdcd_rdo, display_labels = [0, 1])
crdcd_rex.plot()
print("Duration for validation:", crdcd_pt2-crdcd_pt1)
```

```python
crdcd_pt1 = crdcd_fdttee.time()

crdcd_fd_Pp= crdcd_fd_MA.predict(crdcd_fd_i_Xst)
print(crdcd_fdclfrt(crdcd_fd_o_Yst, crdcd_fd_Pp))

crdcd_pt2 = crdcd_fdttee.time()
crdcd_rdo = crdcd_fdcfssm(crdcd_fd_o_Yst, crdcd_fd_Pp)
crdcd_rex = crdcd_fdcfxdp(confusion_matrix = crdcd_rdo, display_labels = [0, 1])
crdcd_rex.plot()
print("Duration for test:", crdcd_pt2-crdcd_pt1)
```
